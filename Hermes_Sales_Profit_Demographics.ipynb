```python
# hermes_dataset.ipynb

# Step 1 — Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For interactive notebook
%matplotlib inline

# Step 2 — Create Base DataFrame for Sales & Profit Metrics
data_financials = {
    'Year': [2022, 2023, 2024, 2025],
    'Revenue_EUR_billion': [12.5, 13.4, 15.17, 16.0],  # estimated / reported figures
    'Gross_Margin_pct': [71.0, 71.1, 70.3, 70.5],   # approximate
    'Operating_Income_pct': [40.0, 42.0, 40.5, 41.0], 
    'Net_Profit_EUR_billion': [3.6, 4.0, 4.6, 4.8]   # approximated
}

df_financials = pd.DataFrame(data_financials)

df_financials.head()
```

```python
# Plot Revenue and Net Profit
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Revenue_EUR_billion', data=df_financials, marker='o', label='Revenue (B€)')
sns.lineplot(x='Year', y='Net_Profit_EUR_billion', data=df_financials, marker='o', label='Net Profit (B€)')
plt.title('Hermès Revenue & Net Profit Over Time')
plt.xlabel('Year')
plt.ylabel('Euros (Billions)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Simulated Demographic Data
data_demo = {
    'Customer_Group': ['Women 18-34', 'Women 35-54', 'Women 55+', 
                       'Men 18-34', 'Men 35-54', 'Men 55+'],
    # Assumed purchase % (simulated)
    'Purchase_Share_pct': [15, 30, 20, 10, 18, 7]
}

df_demo = pd.DataFrame(data_demo)

df_demo['Purchase_Share_pct'] /= df_demo['Purchase_Share_pct'].sum() * 100
df_demo['Purchase_Share_pct'] = df_demo['Purchase_Share_pct'] * 100

df_demo
```

```python
# Demographic donut chart
plt.figure(figsize=(7, 7))
plt.pie(df_demo['Purchase_Share_pct'], labels=df_demo['Customer_Group'], 
        autopct='%1.1f%%', startangle=140, pctdistance=0.75)
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Estimated Hermès Customer Demographic Breakdown')
plt.show()
```

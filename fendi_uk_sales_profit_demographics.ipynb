```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Number of data points
n = 200

# Generate Sales data - annual sales in GBP millions orders
sales = np.round(np.random.normal(loc=15, scale=3, size=n), 2)  # ~£15m mean

# Profit margin (%) between 20% and 60%
profit_margin = np.round(np.random.uniform(20, 60, size=n), 2)

# Demographics
ages = np.random.randint(25, 65, size=n)  # ages 25 to 65
income_k = np.round(np.random.normal(loc=140, scale=35, size=n), 2)  # annual income £k
gender = np.random.choice(["Male","Female","Non‑binary"], size=n, p=[0.44,0.50,0.06])
region = np.random.choice(
    ["London","South East","North West","Midlands","Scotland","Wales","Northern Ireland"],
    size=n
)

# Compile the DataFrame
df = pd.DataFrame({
    "Annual_Sales_£m": sales,
    "Profit_Margin_%": profit_margin,
    "Age": ages,
    "Income_£k": income_k,
    "Gender": gender,
    "Region": region
})

# Display first rows
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
sns.countplot(x="Gender", data=df)
plt.title("Fendi UK Customer Gender Distribution")
plt.show()
```

```python
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution of Fendi UK Customers")
plt.xlabel("Age")
plt.show()
```

```python
sns.histplot(df["Income_£k"], bins=12, kde=True)
plt.title("Income Distribution (£k) of Fendi UK Customers")
plt.xlabel("Annual Income (£k)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Annual_Sales_£m", y="Profit_Margin_%", hue="Region", data=df)
plt.title("Sales vs Profit Margin by Region")
plt.xlabel("Annual Sales (£m)")
plt.ylabel("Profit Margin (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

```python
sns.scatterplot(x="Age", y="Income_£k", data=df)
plt.title("Age vs Income (£k)")
plt.show()
```

```python
df.to_csv("fendi_uk_sales_demographics.csv", index=False)
```

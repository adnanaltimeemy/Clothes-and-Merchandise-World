```python
# ---
# Title: Coach Fashion Label - Sales, Profit Margins & Demographics Dataset
# Author: Your Name
# Description: Build and explore a dataset modeling Coach fashion sales, profit margin, customers
# ---

# 1) IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# 2) DEFINE DATA (Sales, Margins, Demographics)

data = {
    "Year": [2021, 2022, 2023, 2024, 2025, 2026],
    # Annual Total Sales (Coach Brand Segment Only, $ billions) — approximate/estimated
    "Total_Sales": [4.5, 5.0, 5.3, 5.6, 6.0, 6.6],
    # Gross Margin (%) — based on company reporting (2025-2026 ~75-77%) :contentReference[oaicite:1]{index=1}
    "Gross_Margin_pct": [70, 71, 72, 73, 76, 75],
    # Net Profit Margin (%) — assumption for this model
    "Net_Profit_Margin_pct": [8, 9, 9.5, 10, 11.2, 10.8],
    # Customer Age Distribution (approx % by group)
    "Pct_GenZ": [15, 20, 25, 30, 35, 35],
    "Pct_Millennials": [40, 38, 37, 36, 34, 35],
    "Pct_GenX_BabyBoom": [45, 42, 38, 34, 31, 30]
}

# convert to DataFrame
df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe()
```

```python
# Plot Total Sales over Years
plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="Total_Sales", data=df, marker="o")
plt.title("Coach Brand Total Sales ($B) Over Time")
plt.ylabel("Total Sales ($B)")
plt.xlabel("Fiscal Year")
plt.grid(True)
plt.show()

# Plot Margins
plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="Gross_Margin_pct", data=df, marker="o", label="Gross Margin %")
sns.lineplot(x="Year", y="Net_Profit_Margin_pct", data=df, marker="o", label="Net Profit Margin %")
plt.title("Coach Brand Profit Margins Over Time")
plt.ylabel("Margin Percentage")
plt.xlabel("Fiscal Year")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Customer Age Demographics across years
plt.figure(figsize=(12,6))
df_demog = df[["Year","Pct_GenZ","Pct_Millennials","Pct_GenX_BabyBoom"]]
df_demog.plot(x="Year", kind="bar", stacked=True, figsize=(10,6))
plt.title("Coach Brand Customer Age Demographic Trends")
plt.ylabel("Percentage of Customers")
plt.xlabel("Fiscal Year")
plt.legend(title="Age Group")
plt.show()
```

```python
df.to_csv("coach_sales_demographics_dataset.csv", index=False)
print("Dataset saved to coach_sales_demographics_dataset.csv")
```

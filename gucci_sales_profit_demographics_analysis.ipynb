```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display settings
pd.set_option('display.max_columns', 50)
```

```python
# Create revenue/profit margins dataset
gucci_financials = pd.DataFrame({
    'Year': [2022, 2023, 2024, 'H1 2025'],
    'Revenue_EUR_Bn': [10.49, 9.90, 7.65, np.nan],
    'Operating_Margin_%': [36.0, 28.0, 21.0, 16.0]  # approximate figures
})

gucci_financials
```

```python
# ECommerce vs. Retail comparison
sales_channel = pd.DataFrame({
    'Channel': ['Online (gucci.com)', 'Retail Stores'],
    'Sales_USD_Million': [616.6, 7640],  # placeholder for overall retail
    'Percent_of_Total': [616.6 / (616.6 + 7640) * 100, 7640 / (616.6 + 7640) * 100]
})

sales_channel
```

```python
# Demographic distribution
demographics = pd.DataFrame({
    'Age Group': ['18-24', '25-34', '35-44', '45-54', '55-64', '65+'],
    'Percentage': [27.27, 33.30, 18.01, 11.14, 6.53, 3.75]
})

gender_dist = pd.DataFrame({
    'Gender': ['Female', 'Male'],
    'Percentage': [58.84, 41.16]
})

demographics, gender_dist
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Year', y='Revenue_EUR_Bn', data=gucci_financials)
plt.title('Gucci Revenue Over Time (EUR Bn)')
plt.ylabel('Revenue (in â‚¬ billion)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the 4-digit year from values like "H1 2025"
gucci_financials['Year'] = (
    gucci_financials['Year']
        .astype(str)
        .str.extract(r'(\d{4})')[0]
        .astype(int)
)

# Plot
plt.figure(figsize=(8,5))
sns.lineplot(
    x='Year',
    y='Operating_Margin_%',
    data=gucci_financials,
    marker='o'
)

plt.title('Operating Margin Trend')
plt.ylabel('Operating Margin (%)')
plt.xlabel('Year')
plt.xticks(gucci_financials['Year'])  # ensures clean year ticks
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Age Group', y='Percentage', data=demographics)
plt.title('Gucci Customer Age Distribution')
plt.ylabel('Percent of Customers')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(gender_dist['Percentage'], labels=gender_dist['Gender'], autopct='%1.1f%%')
plt.title('Gucci Customer Gender Split')
plt.show()
```

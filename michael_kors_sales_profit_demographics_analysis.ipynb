```python
# -------------------------------------------------------------------
# Step 1  Import core libraries
# -------------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set(style="whitegrid")

# -------------------------------------------------------------------
# Step 2  Create Sample Dataset
# -------------------------------------------------------------------
# Well simulate quarterly sales & profit margin data for 20242025
quarters = [
    "2024_Q1","2024_Q2","2024_Q3","2024_Q4",
    "2025_Q1","2025_Q2","2025_Q3","2025_Q4"
]

data_sales = {
    "Quarter": quarters,
    "Revenue_USD_millions": [800, 760, 780, 790, 810, 750, 770, 815],  # sample
    "Gross_Profit_Margin_pct": [60.5, 59.8, 61.0, 60.3, 59.7, 58.9, 60.1, 60.8],
    "Net_Profit_Margin_pct": [15.0, 14.2, 15.5, 14.8, 14.5, 13.9, 14.2, 15.1]
}

sales_df = pd.DataFrame(data_sales)

# -------------------------------------------------------------------
# Step 3  Create Demographics Dataset
# -------------------------------------------------------------------
# Age, Income, and typical purchase count
demographics_data = {
    "Age_Group": ["18-24","25-34","35-44","45-54","55+"],
    "Pct_Customers": [10, 35, 30, 15, 10],
    "Median_Income_USD": [48000, 72000, 85000, 90000, 78000]
}

demo_df = pd.DataFrame(demographics_data)

# -------------------------------------------------------------------
# Step 4  Display Datasets
# -------------------------------------------------------------------
print("=== Quarterly Sales Data for Michael Kors ===")
display(sales_df)

print("=== Customer Demographics ===")
display(demo_df)

# -------------------------------------------------------------------
# Step 5  Plot Revenue & Margins Over Time
# -------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(sales_df["Quarter"], sales_df["Revenue_USD_millions"], marker="o")
plt.title("Michael Kors Revenue (Sample)")
plt.xlabel("Quarter")
plt.ylabel("Revenue (USD millions)")
plt.xticks(rotation=45)

plt.show()

# -------------------------------------------------------------------
# Step 6  Plot Profit Margins
# -------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(sales_df["Quarter"], sales_df["Gross_Profit_Margin_pct"], label="Gross Margin", marker='o')
plt.plot(sales_df["Quarter"], sales_df["Net_Profit_Margin_pct"], label="Net Margin", marker='o')
plt.title("Profit Margins % Over Time (Sample)")
plt.xlabel("Quarter")
plt.ylabel("Margin (%)")
plt.legend()
plt.xticks(rotation=45)

plt.show()

# -------------------------------------------------------------------
# Step 7  Visualise Demographics Distribution
# -------------------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.barplot(x="Age_Group", y="Pct_Customers", data=demo_df, palette="coolwarm")
plt.title("Customer Age Group Distribution (Sample)")
plt.ylabel("Percentage of Customers")
plt.xlabel("Age Group")

plt.show()

# -------------------------------------------------------------------
# Step 8  Save Dataset to CSV files
# -------------------------------------------------------------------
sales_df.to_csv("michael_kors_sales_data.csv", index=False)
demo_df.to_csv("michael_kors_demographics.csv", index=False)

print("Datasets saved as CSV files.")
```
